```python
%%writefile app.py

import streamlit as st
import pandas as pd

from streamlit_option_menu import option_menu
import sqlite3
conn = sqlite3.connect("mydb.db")
cursor = conn.cursor()
st.title("💰🛢️📈Cross-Market Analysis")

with st.sidebar:
    selected = option_menu("Main Menu", ["Home", 'Filters',"Queries"],
        icons=['house', 'gear'], menu_icon="cast", default_index=1)

if selected == "Home":
    st.write("This is my home page")
if selected == "Queries":
     option = st.selectbox(
    "Select a query",
    ("Display the entire table",
     "Display date",
     "Display price"),
)
     if option == "Display the entire table":
         df = pd.read_sql("select * from oil_prices",con=conn)
         st.dataframe(df)
     if option == "Display date":
         df = pd.read_sql("select date from oil_prices",con=conn)
         st.dataframe(df)




```
